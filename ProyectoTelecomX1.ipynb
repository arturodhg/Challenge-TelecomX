```python
#Ahora me quedan las columnas que tienen categor铆as dentro. Aunque podr铆a asignarles un valor, pienso que ser铆a mejor separar las categor铆as
#en una categor铆a independiente, para que el an谩lisis se pueda realizar mejor.

#Aplico la herramienta get_dummies para dividir las columnas

#InternetService
datos = pd.get_dummies(datos, columns=['InternetService'], prefix=' ', dummy_na=False)

# Contract
datos = pd.get_dummies(datos, columns=['Contract'], prefix=' ', dummy_na=False)

# PaymentMethod
datos = pd.get_dummies(datos, columns=['PaymentMethod'], prefix=' ', dummy_na=False)

#compruebo

datos.info()
```

```python
#Las columnas nuevas que se crearon booleanas, las transformar茅 a tipo entero

for col in datos.columns:
    if datos[col].dtype == bool:
        datos[col] = datos[col].astype(int)

datos.sample(3)
```

```python
datos.info()
```

```python
# En este punto ya todos los datos est谩n estandarizados, pero el DF est谩 desordenado. Lo organizar茅, trasladando las
#columnas de _No, _DSL y _Fiber optic antes de la columna OnlineSecurity y pasar茅 las columnas Charges.Monthtly y Charges.Total al final.

# 1. Agrupo las columnas que quieiro mover
col_grupo1 = [' _No', ' _DSL', ' _Fiber optic']
col_grupo2 = ['Charges.Monthly', 'Charges.Total']

# 2. Obtengo todas las columnas actuales
col_actuales = datos.columns.tolist()

# 3. Inicio la lista del nuevo orden
nuevo_orden = []

#A帽ado las columnas que van antes de 'OnlineSecurity'
for col in col_actuales:
    if col == 'OnlineSecurity':
        break
    if col not in col_grupo1 + col_grupo2:
         nuevo_orden.append(col)

# A帽ado las columnas del grupo de internet
for col_internet in col_grupo1:
    if col_internet in datos.columns:
        nuevo_orden.append(col_internet)

#A帽ado 'OnlineSecurity' y el resto de columnas que van despu茅s
if 'OnlineSecurity' in datos.columns:
     nuevo_orden.append('OnlineSecurity')

     indice_referencia = col_actuales.index('OnlineSecurity')
     for col in col_actuales[indice_referencia + 1:]:
         if col not in col_grupo1 + col_grupo2:
             nuevo_orden.append(col)

# 4. A帽ado las columnas de precios al final
for col_precio in col_grupo2:
    if col_precio in datos.columns:
        nuevo_orden.append(col_precio)


# Verifico
print("\nNuevo orden de columnas:")
print(nuevo_orden)
```

```python
#Guardo los cambios
datos_reorganizado = datos[nuevo_orden].copy()

datos = datos_reorganizado

datos.head()
```

```python
# Para este punto, ya tengo toda la informaci贸n tratada y es 煤til para el an谩lisis. Sin embargo, eliminar茅 la columna
#clientesID y la columna  _No, debido a que contiene datos irrelevantes para el estudio, y cambiar茅 el nombre de las columnas a espa帽ol,
#para hacerlo m谩s entendible.

#primero elimino las columnas irrelevantes.

datos = datos.drop(datos[['customerID',' _No']], axis = 1)
datos.columns.tolist()
```

-Comparativa entre los clientes que cancelaron y los que no cancelaron:

```python
sns.set_theme(style="white")

# Datos para el pie chart
valores = datos['Cancelacion'].value_counts()
etiquetas = ['No Cancelaron', 'Cancelaron']
colores = sns.color_palette("pastel")[:2]  # Colores suaves

# Gr谩fico circular
plt.figure(figsize=(5, 5))
plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', startangle=90, colors=colores, wedgeprops={'edgecolor': 'black'})
plt.title('Porcentajes de clientes que cancelan', fontsize=16)
plt.axis('equal')
plt.tight_layout()

plt.savefig("porcentajes_cancelacion.png", dpi=300, bbox_inches='tight')

plt.show()
```

-Comparaci贸n de cancelaci贸n vs tiempo y gasto mensual y diario:

```python
#Ahora voy a graficar los valores de media, mediana y Desviaci贸n est谩ndar

sns.set_theme(style="whitegrid")  # Cl谩sico limpio con l铆neas
sns.set_theme(style="whitegrid", font_scale=1.1)
colores = sns.color_palette("pastel")[:2]
sns.set_context("notebook", font_scale=1.2)

# Columnas a comparar
cols = ['Meses_contrato', 'Cuentas_mensuales', 'Cuentas_diarias']

nombres_personalizados = {
    'Meses_contrato': 'Meses de contrato',
    'Cuentas_mensuales': 'Pago mensual',
    'Cuentas_diarias': 'Pago diario'
}

# Funci贸n para calcular estad铆sticas por grupo
def resumen_estadisticas(datos, columnas):
    lista = []
    for col in columnas:
        cancelaron = datos[datos['Cancelacion'] == 1][col]
        no_cancelaron = datos[datos['Cancelacion'] == 0][col]

        lista.append(['Media', col, 'Cancelaron', cancelaron.mean()])
        lista.append(['Media', col, 'No Cancelaron', no_cancelaron.mean()])
        lista.append(['Mediana', col, 'Cancelaron', cancelaron.median()])
        lista.append(['Mediana', col, 'No Cancelaron', no_cancelaron.median()])
        lista.append(['Desviaci贸n est谩ndar', col, 'Cancelaron', cancelaron.std()])
        lista.append(['Desviaci贸n est谩ndar', col, 'No Cancelaron', no_cancelaron.std()])

    df = pd.DataFrame(lista, columns=['Estadistica', 'Variable', 'Grupo', 'Valor'])
    return df

# Calculo el resumen
df_resumen = resumen_estadisticas(datos, cols)

# Crear gr谩fico con subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 7), sharey=False)

# Gr谩ficos por estad铆stica
estadisticas = ['Media', 'Mediana', 'Desviaci贸n est谩ndar']
colores = {'Cancelaron': 'tomato', 'No Cancelaron': 'skyblue'}

for i, estadistica in enumerate(estadisticas):
    df_plot = df_resumen[df_resumen['Estadistica'] == estadistica].copy()
    df_plot['Variable'] = df_plot['Variable'].map(nombres_personalizados)

    sns.barplot(data=df_plot, x='Variable', y='Valor', hue='Grupo',
                palette=colores, ax=axes[i])

    axes[i].set_title(estadistica, fontsize = 18, loc ='left')


    # Ajusto el eje Y seg煤n el tipo de estad铆stica
    if estadistica == 'Media':
        axes[i].set_ylim(0, 90)
    elif estadistica == 'Mediana':
        axes[i].set_ylim(0, 90)
    elif estadistica == 'Desviaci贸n est谩ndar':
        axes[i].set_ylim(0, 35)

    axes[i].set_ylabel('',)
    axes[i].set_xlabel('')
    axes[i].tick_params(axis='x',size = 14)
    axes[i].legend(title = '',loc='upper right')

    # Mostrar los valores encima de las barras
    for container in axes[i].containers:
        axes[i].bar_label(container, fmt='%.2f%%', label_type='edge', padding=3, fontsize = 10)

# T铆tulo general y ajuste del espacio superior
plt.suptitle('Comparativa de los porcentajes de Cancelaci贸n de clientes en relaci贸n con \nel tiempo de permanencia y los pagos mensuales y diarios', fontsize=20)
plt.subplots_adjust(top=0.85)
plt.tight_layout()

plt.savefig("comparacion_pago_mes_dia_y_tiempo.png", dpi=300, bbox_inches='tight')


plt.show()
```

-Con esta 煤ltima gr谩fica poremos observar que:

 *Los clientes que se fueron, ten铆an un porcentaje de cobro mensual y duario, m谩s elevado que el de los clientes que se quedaron.

Comparativa de cancelaci贸n por tipo de servicios

```python
#primero creo una lista con las columnas que quiero comparar
#En este caso, la relativas a servicios
servicios = ['Servicio_telefonico', 'Multiples_lineas', 'DSL', 'Fibra_optica']
```

```python
#ahora lo grafico

sns.set_theme(style="whitegrid")

# Servicios a analizar
nombres_barras = {
    'Servicio_telefonico': 'Tel茅fono',
    'Multiples_lineas': 'M煤ltiples L铆neas',
    'DSL': 'Internet DSL',
    'Fibra_optica': 'Fibra 贸ptica'
}

# Preparar DataFrame con proporciones por grupo
df_resultados = []

for servicio in servicios:
    for grupo in [0, 1]:  # 0 = No Cancelaron, 1 = Cancelaron
        subset = datos[datos['Cancelacion'] == grupo]
        total = len(subset)
        con_servicio = subset[servicio].sum()
        proporcion = (con_servicio / total) * 100 if total > 0 else 0  #  en porcentaje

        df_resultados.append([
            nombres_barras[servicio],
            'Cancelaron' if grupo == 1 else 'No Cancelaron',
            proporcion
        ])

df_servicios = pd.DataFrame(df_resultados, columns=['Servicio', 'Grupo', 'Proporcion'])

# Grafico
plt.figure(figsize=(10, 6))

sns.barplot(
    data=df_servicios,
    x='Servicio',
    y='Proporcion',
    hue='Grupo',
    palette='pastel',
    errorbar=None
)

plt.title('Proporci贸n de clientes con servicio seg煤n grupo de cancelaci贸n', fontsize=16)
plt.xlabel('Servicio')
plt.ylabel('Proporci贸n con servicio (%)')
plt.ylim(0, 100)

# Etiquetas de porcentaje sobre cada barra
for container in plt.gca().containers:
    plt.bar_label(container, fmt='%.0f%%', label_type='edge', padding=3)

plt.tight_layout()
plt.show()
```

Aqui podemos ver que:

*Con respecto a las l铆neas m煤ltiples hay una ligera relaci贸n de cancelaci贸n por parte de los clientes.

*La proporci贸n de clientes que se fuern es mayor con respecto al internet de Fibra 贸ptica.

*Contrario al Servicio de DSL cuyo porcentaje es ligeramente mayor en los clientes que se quedaron.
